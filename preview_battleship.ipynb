```python
import numpy as np
```

```python
# Iniciamos un tablero de 10 x 10 
board = np.full((10,10)," ")

# Señalamos número de barcos en tablero
n = 1

# Inicializamos contador de barcos
n_boats = 0

# Creamos n barcos aleatorios
while True:
    barco_1 = np.full((2, 1), "O")
    barco_2 = np.full((1, 4), "O")
    rng1 = np.random.randint(0, 8)
    rng2 = np.random.randint(0, 8)
    orientation = np.random.randint(1,3)
    size_until_border = board[rng1:rng1+1, rng2-1:-1]
    size_until_border_2 = board[rng1-1:-1, rng2:rng2+1]
    if orientation == 1:
        while (board[rng1, rng2] != "O") & (len(barco_2[0])<=len(size_until_border[0])) & ("O" not in size_until_border):
            board[rng1:rng1+1,rng2:len(barco_2[0])+rng2] = barco_2
            n_boats+=1
            break
    elif orientation == 2:
        while (board[rng1, rng2] != "O") & (len(barco_1)<=len(size_until_border_2)) & ("O" not in size_until_border_2):
            board[rng1:rng1+len(barco_1),rng2:rng2+1] = barco_1
            n_boats+=1
            break
    if n_boats == n:
        break

# Iniciamos función de disparo
while True:
    print(board) # Print chivato para pruebas
    x_cordinate = int(input("Inserte coordenada x"))
    y_cordinate = int(input("Inserte coordenada y"))
    if board[x_cordinate, y_cordinate] == "O":
        board[x_cordinate, y_cordinate] = "X"
    else:
        board[x_cordinate,y_cordinate] = "-"
        print ("Agua")
    if ("O" in board) == False:
        print("¡Felicidades, has ganado!")
        break

print(board)
```
